```python
import os
from rdkit import Chem
from rdkit.Chem import AllChem

smile_dict = {
  'PSPCz': 'CC1=CC=C(C=C1)S(=O)(=O)C2=CC=CC3=C2NC4=CC=CC=C34',
  '4CzIPN': 'CC1=CC=C2C(=C1)C3=CC=CC=C3N2C4=C(C(=C(C(=C4C#N)N5C6=CC=CC=C6C7=CC=CC=C75)N8C9=CC=CC=C9C1=CC=CC=C18)N1C2=CC=CC=C2C2=CC=CC=C21)C#N',
  'DMAC1': 'CC1(C2=CC=CC=C2NC3=CC=CC=C31)C',
  'DMAC2': 'CC1=CC=C2C(=C1)NC3=CC=CC=C3O2',
  'AZB1': 'B1=CC=NC=C1'
}

output_dir = 'data'
if not os.path.exists(output_dir):
  os.makedirs(output_dir)

for key, smile in smile_dict.items():
  mol = Chem.MolFromSmiles(smile)

  if mol is None:
    print(f"Erreur SMILES pour {key}")
    continue

  Chem.SanitizeMol(mol)
  Chem.Kekulize(mol)

  mol = Chem.AddHs(mol)
  AllChem.EmbedMolecule(mol)

  AllChem.UFFOptimizeMolecule(mol, maxIters=200)

  conf = mol.GetConformer()
  for i in range(conf.GetNumAtoms()):
    pos = conf.GetAtomPosition(i)
    print(pos)

  with open(os.path.join(output_dir, f"{key}.xyz"), 'w+') as f:
    f.write(str(mol.GetNumAtoms()) + '\n\n')

    for i in range(mol.GetNumAtoms()):
      atom = mol.GetAtomWithIdx(i)
      pos = conf.GetAtomPosition(i)
      f.write(f'{atom.GetSymbol()} {pos.x:.3f} {pos.y:.3f} {pos.z:.3f}\n')

  print(f"Fichier {key}.xyz généré")
```
